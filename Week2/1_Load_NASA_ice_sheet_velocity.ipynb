```python
import xarray as xr
```

```python
url1 = 'http://its-live-data.s3.amazonaws.com/datacubes/v2/S50W030/ITS_LIVE_vel_EPSG32724_G0120_X550000_Y3950000.zarr'
url2 = 'http://its-live-data.s3.amazonaws.com/datacubes/v2/S50W030/ITS_LIVE_vel_EPSG32724_G0120_X550000_Y4050000.zarr'
url3 = 'http://its-live-data.s3.amazonaws.com/datacubes/v2/S50W030/ITS_LIVE_vel_EPSG32724_G0120_X650000_Y3950000.zarr'
url4 = 'http://its-live-data.s3.amazonaws.com/datacubes/v2/S50W030/ITS_LIVE_vel_EPSG32724_G0120_X650000_Y4050000.zarr'
url5 = 'http://its-live-data.s3.amazonaws.com/datacubes/v2/S50W030/ITS_LIVE_vel_EPSG32725_G0120_X350000_Y3950000.zarr'
```

```python
ds1 = xr.open_dataset(url1, engine= 'zarr', chunks="auto")['v']
ds1.sortby('mid_date')
```

```python
ds2 = xr.open_dataset(url2, engine= 'zarr', chunks="auto")['v']
ds2.sortby('mid_date')
```

```python
ds3 = xr.open_dataset(url3, engine= 'zarr', chunks="auto")['v']
ds3.sortby('mid_date')
```

```python
ds4 = xr.open_dataset(url4, engine= 'zarr', chunks="auto")['v']
ds4.sortby('mid_date')
```

```python
ds5 = xr.open_dataset(url5, engine= 'zarr', chunks="auto")['v']
ds5.sortby('mid_date')
```

```python
import matplotlib.pyplot as plt
```

```python
k1 = ds1.isel(mid_date = 1).load()
k2 = ds2.isel(mid_date = 1).load()
k3 = ds3.isel(mid_date = 1).load()
k4 = ds4.isel(mid_date = 1).load()
k5 = ds5.isel(mid_date = 1).load()
```

```python
xr.merge([k1.drop('mid_date'), k2.drop('mid_date'), k3.drop('mid_date'), k4.drop('mid_date'), k5.drop('mid_date')])['v'].plot()
```
